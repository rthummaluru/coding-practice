### Lets first explore columns with null values

```python
# Great example to look at ontime and delay as meaningful missing data opposed to these other columns 

na_dropped_df = df.drop(['Driver_Name', 'Driver_MobileNo', 'Minimum_kms_to_be_covered_in_a_day'], axis=1)
na_dropped_df
```

```python
# check all columns
pd.set_option('display.max_columns', None)
na_dropped_df
```

notice some important fields such as planned_eta vs actual eta, transportation_distance_in_km, ontime/delay

```python
# change the values of ontime and delay to binary 1's and 0's 
na_dropped_df['ontime'] = na_dropped_df['ontime'].replace({'G' : 1})
na_dropped_df['ontime'] = na_dropped_df['ontime'].fillna(0)

na_dropped_df['delay'] = na_dropped_df['delay'].replace({'R' : 1})
na_dropped_df['delay'] = na_dropped_df['delay'].fillna(0)
na_dropped_df
```

```python
# format the headings
na_dropped_df.columns = na_dropped_df.columns.str.strip().str.lower().str.replace(' ', '_')

# split latitude and longitutde
na_dropped_df[['org_lat', 'org_lon']] = na_dropped_df['org_lat_lon'].str.split(',', expand=True).astype(float)
na_dropped_df[['des_lat', 'des_lon']] = na_dropped_df['des_lat_lon'].str.split(',', expand=True).astype(float)
```
